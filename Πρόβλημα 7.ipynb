```python
import numpy as np
def U() : return np.random.rand()
import matplotlib.pyplot as plt
```

```python
N_0 = 1000 
N = []
N += [N_0]
Δt = 1
p = 10**-3 #πιθανότητα διάσπασης
T = 4000

for i in range(T):
    dN = 0
    for i in range(N_0):
        if U() < p : dN = -1
    N_0+=dN
    N+= [N_0]
plt.plot(N,color='purple',label='Πλήθος αδιάσπαστων πυρήνων')
plt.xlabel('T')
plt.ylabel('N')
plt.legend()
```
