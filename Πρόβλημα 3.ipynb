```python
import numpy as np
import matplotlib.pyplot as plt
def U(): return np.random.rand()
```

```python
def roll_dice():
    random_number = U()
    random_number=int(random_number*6 + 1)
    return random_number

dice = [roll_dice() for _ in range(10000)]
plt.hist(dice,bins=range(0,8),ec='black')
```

```python
#άλλος τρόπος
def dice():
    res = 6
    x = U()*6
    if x<5: res = 5
    if x<4: res = 4
    if x<3: res = 3
    if x<2: res = 2
    if x<1: res = 1
    return res

dices = [dice() for _ in range(1000)]
plt.hist(dices, bins = range(0,8), ec='black')
```

```python
def unfair_dice():
    result = 0
    x = U()
    if x < 0.3: result = 6
    if 0.3 < x < 0.44 : result = 5
    if 0.44 < x < 0.58 : result = 4
    if 0.58 < x < 0.72 : result = 3
    if 0.72 < x < 0.84 : result = 2
    if 0.84 < x < 1 : result = 1
    return result
    
dices = [unfair_dice() for i in range(10000)]
plt.hist(dices,bins=range(0,8),ec='black')
```
