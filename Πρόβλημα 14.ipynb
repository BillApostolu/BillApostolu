```python
import numpy as np
import math
import matplotlib.pyplot as plt 
```

```python
def f(x): return np.tan(x) - x/(1 - x**2)
a, b = 2 , 4
x = np.linspace(a, b, 1000)
y = f(x)
plt.plot(x, y)
plt.axhline(y=0., color='green', linestyle='-')
```

```python
def dixotomisi(f, a, b, e):
    steps = []
    while(abs(b - a) > e):
        xnew = (a + b)/2
        steps.append(xnew)
        if f(xnew)*f(b) < 0: a = xnew
        else: b = xnew
        print('iteration = ', len(steps), ': [', a,',',b,']')
    return np.array(steps)

a, b, e = 2 , 4, 1e-4
points = dixotomisi(f, a, b, e)
print("Η ρίζα είναι: ρ = %2.3f" %points[-1], 'με f(ρ) = ',f(points[-1]))
```

```python
def f_der(x): return 1/(np.cos(x))**2 + 1/(x**2 - 1) - (2*x**2)/(x**2 - 1)**2
```

```python
def newton_raphson(x1, e, f, f_der):  
    x, xprev = x1 , x1-2*e  
    steps = []

    while(abs(x - xprev) > e): 
        xprev = x
        x = x - f(x)/f_der(x) 
        steps.append(x) 
        print('iteration = ', len(steps), ': [', xprev,',',x,']')
                    
    return np.array(steps)
```

```python
x1, e = 3, 1e-10
points=newton_raphson(x1, e , f ,f_der)
print("Η ρίζα είναι: ρ = %2.3f" %points[-1], 'με f(ρ) = ',f(points[-1]))
```
