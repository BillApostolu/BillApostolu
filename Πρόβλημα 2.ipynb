```python
c = 2147483629
a = 2147483587
Nmax = 2**31 - 1
x = 68 #seed
random_numbers = []

for i in range(100000):
    x = (c*x+a)%Nmax
    random_number = x/Nmax 
    random_numbers.append(random_number) #The append() method appends an element to the end of the list: list.append(elmnt)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.hist(random_numbers,bins=100,ec='black')
```
