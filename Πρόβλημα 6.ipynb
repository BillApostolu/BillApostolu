```python
import numpy as np 
import matplotlib.pyplot as plt
def step(): 
    l = 1
    if np.random.rand() < 0.5: l = -1
    return l
#Για Μ=1 σωματίδια
N  = 10000 # αριθμός βημάτων
path = []
x  = 0
path += [x] 
for n in range(N):
    x += step()
    path = path + [x]

print('x = %d'%(x))
plt.plot(range(len(path)), path)
print(path[0:10])
```

```python
import random
import math

def brownian_motion(num_particles, num_steps, step_size):
    # Αρχικοποίηση των συντεταγμένων των σωματιδίων
    particles = [(0, 0) for _ in range(num_particles)]
    
    # Προσομοίωση της κίνησης Brownian για κάθε βήμα
    for _ in range(num_steps):
        for i in range(num_particles):
            # Επιλογή τυχαίας κατεύθυνσης
            direction = random.choice([(0, 1), (0, -1), (1, 0), (-1, 0)])
            # Κίνηση του σωματιδίου στην επιλεγμένη κατεύθυνση με το step_size
            particles[i] = (particles[i][0] + direction[0] * step_size, particles[i][1] + direction[1] * step_size)
    
    # Υπολογισμός της απόστασης από την αρχή των αξόνων για κάθε σωματίδιο
    distances = [math.sqrt(x**2 + y**2) for x, y in particles]
    
    # Υπολογισμός του μέσου όρου των αποστάσεων
    mean_distance = sum(distances) / len(distances)
    
    return mean_distance

num_particles = 1000
num_steps = 100
step_size = 1  # Η ταχύτητα είναι 1 cm/s

mean_distance = brownian_motion(num_particles, num_steps, step_size)
print("Ο μέσος όρος της απόστασης μετά από 100 βήματα είναι:", mean_distance)
plt.hist(distances, bins=range(-100,100))
```
