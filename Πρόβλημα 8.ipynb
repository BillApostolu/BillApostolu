```python
import numpy as np
import matplotlib.pyplot as plt
def U(): return np.random.rand()
```

```python
N1_0, N2_0 = 100 , 0 
N1 = []
N2 = []
N1 += [N1_0]
N2 += [N2_0]
p=10**-2

for i in range(500):
    dN1, dN2 = 0, 0
    
    for i in range(N1_0):
        if U() < p : dN1 -= 1
    for i in range(N2_0):
        if U() < p : dN2 +=1
            
    N1_0 +=  dN1 + dN2
    N2_0 += - dN1 - dN2
    
    N1+=[N1_0]
    N2+=[N2_0]
    
plt.plot(N1,color='blue',label=' 1 ')
plt.plot(N2,color='green',label=' 2 ')
plt.xlabel('t')
plt.ylabel('1, 2')
plt.legend()
```
