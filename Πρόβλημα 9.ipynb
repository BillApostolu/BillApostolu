```python
import numpy as np
import matplotlib.pyplot as plt
def U(): return np.random.rand()
```

```python
S1 = []
S2 = []
S20 = []

for i in range(10000):
    x1 = [U()]
    mesos_oros_S1 = np.mean(x1)
    S1 += [mesos_oros_S1]
    std_S1 = np.std(x1)
    x2 = [U() for i in range(2)]
    mesos_oros_S2 = np.mean(x2)
    S2 += [mesos_oros_S2]
    std_S2 = np.std(x2)
    x20 = [U() for i in range(20)]
    mesos_oros_S20 = np.mean(x20)
    S20 += [mesos_oros_S20]
    std_S20 = np.std(x20)
print('Ο μέσος όρος και η τυπική απόκλιση του δείγματος S1 είναι',mesos_oros_S1, 'και',std_S1, 'αντίστοιχα.')
print('Ο μέσος όρος και η τυπική απόκλιση του δείγματος S2 είναι',mesos_oros_S2, 'και',std_S2, 'αντίστοιχα.')
print('Ο μέσος όρος και η τυπική απόκλιση του δείγματος S20 είναι',mesos_oros_S20, 'και',std_S20, 'αντίστοιχα.')
```

```python
plt.hist(S1 , bins = np.linspace(0, 1), ec = 'black')
plt.hist(S2 , bins = np.linspace(0, 1), ec = 'black')
plt.hist(S20, bins = np.linspace(0, 1), ec = 'black')
```
