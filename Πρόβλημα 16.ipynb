```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def f(x): return x-5-np.sqrt(x)
def g(x): return 5 + np.sqrt(x)
a, b = 0 , 10
x = np.linspace(a, b, 1000)
y = g(x)
Y = x
plt.plot(x, y)
plt.plot(x, Y, color='green', linestyle='-')
```

```python
def fixed_points(x0, g, e):
    x , xprev = x0, x0+2*e
    steps = []
    
    while(abs(xprev - x)> e):
        xprev = x
        x = g(x)
        steps.append(x)
        print('iteration = ', len(steps), ': [', xprev,',',x,']')
    return np.array(steps)
```

```python
points = fixed_points(4, g, 10**(-6))
print("Η ρίζα είναι: ρ = %2.3f" %points[-1], 'με f(ρ) = ',f(points[-1]))
```
