```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import display
```

```python
def der_A(lamda, t, A):
    return -lamda*A

def rungekutta4(a, b, A0, lamda, dt):
    N = int((b-a)/dt)
    t , A = np.zeros(N) , np.zeros(N)
    t[0] = a
    A[0] = A0
    for i in range(N-1):
        k1 = dt*der_A(lamda, t[i], A[i])
        k2 = dt*der_A(lamda, t[i] + dt/2, A[i] + dt/2*k1)
        k3 = dt*der_A(lamda, t[i] + dt/2, A[i] + dt/2*k2)
        k4 = dt*der_A(lamda, t[i] + dt, A[i] + dt*k3)
        A[i+1] = A[i] + (1/6)*(k1 + 2*k2 + 2*k3 +k4)
        t[i+1] = t[i] + dt
    return A , t

#Τυχαίες τιμ΄ές
x = np.linspace(0,8,100)
A_th = 1*np.exp(-(np.log(2)/2)*x)

plt.xlabel("t [s]")
plt.ylabel("A [kBq]")
plt.title("Μέθοδος Runge - Kutta Τέταρτης Τάξης")
energothta, time = rungekutta4(0, 8, 1, np.log(2)/2, 0.1)
plt.scatter(energothta, time, marker = '.', color='b', label='Μέθοδος Runge Kutta 4ης Τάξης για βήμα 0.1')
plt.plot(A_th, x, 'black', label = 'Αναλυτική Καμπύλη')

plt.legend()
```
