```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 + 1
def fdot(x):
    return 2*x
x = np.linspace(1,3,100)
y = f(x)
plt.title('Γραφική παράσταση f(x)')
plt.plot(x,y,color='red')
plt.grid(True)
```
