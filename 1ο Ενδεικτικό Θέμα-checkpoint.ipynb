```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([ 0 , 1 , 2 , 3 , 4])
y = np.array([ 0 , 0 , 8 , 54 , 192])
```

```python
def lagrange_secondorder(x, x_data, y_data, pos):
    N = len(pos)
    p2 = 0 
    for i in range(N-2):
        ff0 = (y_data[pos[i]] * (x - x_data[pos[i+1]]) * (x - x_data[pos[i+2]])) / ((x_data[pos[i]] - x_data[pos[i+1]]) * (x_data[pos[i]] - x_data[pos[i+2]]))
        ff1 = (y_data[pos[i+1]] * (x - x_data[pos[i]]) * (x - x_data[pos[i+2]])) / ((x_data[pos[i+1]] - x_data[pos[i]]) * (x_data[pos[i+1]] - x_data[pos[i+2]]))
        ff2 = (y_data[pos[i+2]] * (x - x_data[pos[i]]) * (x - x_data[pos[i+1]])) / ((x_data[pos[i+2]] - x_data[pos[i]]) * (x_data[pos[i+2]] - x_data[pos[i+1]]))
        p2 += ff0 + ff1 + ff2
    return p2

pos = [0, 1, 2]

x_values = np.linspace(0, 4, 100)
y_values = [lagrange_secondorder(x_val, x, y, pos)  for x_val in x_values]

plt.scatter(x, y, label='Σημεία Δεδομένων')
plt.plot(x_values, y_values, label='Παρεμβολή Lagrange Δεύτερης Τάξης', color='red')

plt.legend()
plt.grid(True)

print(f"Για x = 1.5, από παρεμβολή 2ης τάξης: f(1.5) =", lagrange_secondorder(1.5, x, y, [0, 1, 2]))
```

```python
def newton_gregory(x_data, f_data, h):
    N = len(f_data)
    for i in range(N):
        if i + 2 < N:
            rate = - (f_data[i+2] - 4*f_data[i+1] + 3*f_data[i]) / (2*h)
            print(f'Για x = {x_data[i]}, από 2ης τάξης προσέγγιση παραγώγου f\'({x_data[i]}) = {round(rate, 6)}')
        else:
            rate1 = (f_data[i-2] - 4*f_data[i-1] + 3*f_data[i]) / (2*h)
            print(f'Για x = {x_data[i]}, από 2ης τάξης προσέγγιση παραγώγου f\'({x_data[i]}) = {round(rate1, 6)}')

newton_gregory(x, y, 1)
```

```python
def simpson(y_data, a, b, N):
    h = (b-a)/N
    int2 , int3 = 0 , 0
    int1 = y_data[a] + y_data[b]
    for i in range(N):
        if i%2 == 1:
            k = a + i*h
            if 0 <= k <= 2:
                pos = [0 , 1 , 2]
            else:
                pos = [2 , 3 , 4]
            int2 += 4 * lagrange_secondorder(k, x, y_data, pos)    
        else:
            k = a + i*h
            if 0 <= k <= 2:
                pos = [0 , 1 , 2]
            else:
                pos = [2 , 3 , 4]
            int3 += 2 * lagrange_secondorder(k, x, y_data, pos)    
    integral = (h/3)*(int1 + int2 + int3)
    return integral

print("Η τιμή του ολοκληρώματος με τη γενικευμένη μέθοδο Simpson είναι:", simpson(y, 0, 2, 1000) + simpson(y, 2, 4, 1000))
```
